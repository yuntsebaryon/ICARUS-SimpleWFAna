```python
import ROOT
```

```python
fsin = ROOT.TF1("fsin", "sin(x)+sin(2*x)+sin(0.5*x)", 0, 4*ROOT.TMath.Pi())
```

```python
c1 = ROOT.TCanvas("fsin", "fsin", 1600, 1200)
fsin.Draw()
c1.Draw()
```

```python
n = 25
hsin = ROOT.TH1D("hsin", "hsin", n+1, 0, 4*ROOT.TMath.Pi())
for i in xrange( 0, n ):
    x = ( float(i)/n)*(4*ROOT.TMath.Pi())
    hsin.SetBinContent( i+1, fsin.Eval(x) );
```

```python
hsin.Draw("same")
c1.Draw()
```

```python
hm = ROOT.TH1D()
ROOT.TVirtualFFT.SetTransform(0)
hm = hsin.FFT(hm, "MAG")
```

```python
c2 = ROOT.TCanvas("hm", "hm", 1600, 1200)
hm.Draw()
c2.Draw()
```
